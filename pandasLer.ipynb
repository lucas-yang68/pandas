```python


import pandas as pd
import numpy as np

def adder(ele1,ele2):
   return ele1+ele2

df = pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])
df.pipe(adder,2)
print(df)
print(df.apply(np.mean))
print(np.random.randn(2,3))
```

```python
import pandas as pd
import numpy as np

df1 = pd.DataFrame(np.random.randn(10,4),columns=['col1','col2','a','b'])
df2 = pd.DataFrame(np.random.randn(7,3),columns=['col1','col2','col3'])
print(df1)
df1 = df1.reindex_like(df2)
print(df1)
```

```python
labels = ['A', 'B', 'C', 'D','C']
index_map = {k: i for i, k in enumerate(labels) if k not in labels[:i]}
print(index_map['C'])  # 输出：2
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(10, 4),
   index = pd.date_range('1/1/2000', periods=10),
   columns = ['A', 'B', 'C', 'D'])

print(df)
r = df.rolling(window=3,min_periods=1)
print(r)
```
